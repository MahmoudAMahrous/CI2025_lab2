the traveling salesman problem
we need to find the shortest route that passes by the cities once and returns to the starting point

```python
import numpy as np
import random
from icecream import ic
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
data = []
populationSize = 10

# in this list i am loading each problem and considering the number of population and the max iterations
problems = [['test_problem.npy', 10, 10000, 0.3],
            ['problem_g_10.npy', 10, 10000, 0.2],
            ['problem_g_20.npy', 40, 10000, 0.2],
            ['problem_g_50.npy', 80, 10000, 0.2],
            ['problem_g_100.npy', 110, 10000, 0.2],
            ['problem_g_200.npy', 140, 10000, 0.2],
            ['problem_g_500.npy', 170, 10000, 0.2],
            ['problem_g_1000.npy', 200, 10000, 0.2],
            ['problem_r1_10.npy', 10, 10000, 0.2],
            ['problem_r1_20.npy', 40, 10000, 0.2],
            ['problem_r1_50.npy', 80, 10000, 0.2],
            ['problem_r1_100.npy', 110, 10000, 0.2],
            ['problem_r1_200.npy', 140, 10000, 0.2],
            ['problem_r1_500.npy', 170, 10000, 0.2],
            ['problem_r1_1000.npy', 200, 10000, 0.2],
            ['problem_r2_10.npy', 10, 10000, 0.2],
            ['problem_r2_20.npy', 40, 10000, 0.2],
            ['problem_r2_50.npy', 80, 10000, 0.2],
            ['problem_r2_100.npy', 110, 10000, 0.2],
            ['problem_r2_200.npy', 140, 10000, 0.2],
            ['problem_r2_500.npy', 170, 10000, 0.2],
            ['problem_r2_1000.npy', 200, 10000, 0.2]]


def randomSolution():
    numbers = list(range(problemSize))
    random.shuffle(numbers)
    return numbers


def routeLength(solution):
    result = 0
    for i in range(problemSize):
        result += data[solution[i]][solution[(i+1) % problemSize]]
    return result


def orderBasedOnFitness(solutionsList):
    fitnesses = [routeLength(s) for s in solutionsList]
    solutionsWithFitnesses = list(zip(solutionsList, fitnesses))
    orderedSolutions = sorted(solutionsWithFitnesses,
                              key=lambda x: x[1])
    return [s for s, _ in orderedSolutions]


def crossOver(parent1, parent2):
    start, end = sorted(random.sample(range(problemSize), 2))
    child = [None] * problemSize
    child[start:end+1] = parent1[start:end+1]
    pointer = 0
    for city in parent2:
        if city not in child:
            while child[pointer] is not None:
                pointer += 1
            child[pointer] = city

    return child


def swapMutation(solution):
    newSolution = solution[:]
    a, b = random.sample(range(len(solution)), 2)
    newSolution[a], newSolution[b] = newSolution[b], newSolution[a]
    return newSolution


def tournamentSelection(population, nSelection):
    selections = []
    for _ in range(nSelection):
        selections.append(random.choice(population))
    return orderBasedOnFitness(selections)[0]


def solve(maxSteps, mutationProbability=0.2):
    # generate population
    population = [randomSolution() for _ in range(populationSize)]

    history = []
    bestDistance = 10**50
    bestSolution = []
    for _ in tqdm(range(maxSteps)):
        newChildren = []
        for _ in range(round(populationSize / 5)):
            if (random.random() < mutationProbability):
                newChildren.append(swapMutation(
                    tournamentSelection(population, 4)))
            else:
                newChildren.append(crossOver(tournamentSelection(population, 4),
                                             tournamentSelection(population, 4)))
        population.extend(newChildren)
        population = orderBasedOnFitness(population)
        population = population[0:populationSize]
        iterationBestDistance = routeLength(population[0])
        if (iterationBestDistance < bestDistance):
            bestDistance = iterationBestDistance
            bestSolution = population[0]
        history.append(iterationBestDistance)

    return bestDistance, bestSolution, history


problemsBestDistances = []
for problem in problems:
    data = np.load('problems/' + problem[0])
    problemSize = len(data)
    populationSize = problem[1]
    bestDistance, bestSolution, history = solve(problem[2], problem[3])
    problemsBestDistances.append(bestDistance)
    ic(problem[0], bestSolution, bestDistance)
    plt.plot(history)
    plt.title("lowest distance over time")
    plt.show()
ic(problemsBestDistances)
```
